##DESAFIO AULA 02

###Desafio 01: Ordernar o grÃ¡fico countplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
plt.figure(figsize=(10, 8))

ax = sns.countplot(x = 'composto', data=dados.query('composto in @cod_composto'), order=cod_composto)
ax.set_title('Top 5 Compostos')
plt.show()
```

###Desafio 02: Melhorar a visualizaÃ§Ã£o alterando tamanho da fonte

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
plt.figure(figsize=(10, 8))

ax = sns.countplot(x = 'composto', data=dados.query('composto in @cod_composto'), order=cod_composto, hatch ='-')
plt.xlabel('Composto', size=15)
plt.ylabel('Quantidade', size=15)
plt.tight_layout()
ax.set_title('Top 5 Compostos Classificado por Quantidade', size=20)
plt.show()
```
